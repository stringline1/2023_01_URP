```python
import nltk
from konlpy.tag import Hannanum

han = Hannanum()

#품사 태깅
def extract_words(sent):
  try:
    tagging = han.pos(sent)
    word_list = []
    for i in range(0, len(tagging)):
        if tagging[i][1] == 'N' or  tagging[i][1] == 'P': #체언&용언
            word_list.append(tagging[i][0])

    return word_list
  except:
    return sent

#상담 데이터 품사 태깅
dsm['pos'] = dsm['word'].apply(extract_words)

dsm
```

```python
import pandas as pd

depressed = pd.read_csv('/content/drive/MyDrive/학연생/데이터/depressed_pos.csv', encoding='UTF8')
daily = pd.read_csv('/content/drive/MyDrive/학연생/데이터/일상_pos.csv', encoding='UTF8')
```

우울장애 유사도

```python
# word2vec 모델 학습에 로그를 찍을 수 있도록 합니다.
import logging
logging.basicConfig(
    format='%(asctime)s : %(levelname)s : %(message)s', 
    level=logging.INFO)
```

```python
def to_list(sent):
  try:
    return eval(sent)
  except:
    return None

depressed['pos'] = depressed['pos'].apply(to_list)

depressed = depressed.dropna()

depressed['pos']
```

```python
from soynlp.tokenizer import RegexTokenizer

tokenizer = RegexTokenizer()

#토큰화
depressed['pos'] = depressed['pos'].apply(' '.join)

tokens = depressed['pos'].apply(tokenizer.tokenize)
depressed['tokenized'] = depressed['pos'].apply(tokenizer.tokenize)

depressed['tokenized']
```

```python
# 초기화 및 모델 학습
from gensim.models import word2vec

# 모델 학습
model = word2vec.Word2Vec(tokens, min_count=1)

# 모델 이름을 지정하고 저장한다.
model_name = 'depressed'
model.save(model_name)
```

```python
def similar(word):
  try:
    for w in word:
      words = []
      for i,j in model.wv.most_similar(w):
        words.append(i)
    return words
  except:
    return None

dsm['similar'] = dsm['pos'].apply(similar)
dsm
```

```python
from google.colab import files
dsm.to_csv('dsm_우울.csv', encoding = 'utf-8-sig') 
files.download('dsm_우울.csv')
```
