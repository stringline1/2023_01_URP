```python
import re

def preprocessing(text):
    text = re.sub('\\\\n', ' ', text)
    text = re.sub('[?.,;:|\)*~`’!^\-_+<>@\#$%&-=#}※]', '', text)
    text = re.sub('[^가-힣ㄱ-ㅎㅏ-ㅣa-zA-Z]', ' ', text)
    return text


depressed['utterance'] = depressed['utterance'].apply(preprocessing)

depressed['utterance']
```

DSM 형태소 분석

```python
import nltk
from konlpy.tag import Hannanum

han = Hannanum()

#품사 태깅
def extract_words(sent):
  try:
    tagging = han.pos(sent)
    word_list = []
    for i in range(0, len(tagging)):
        if tagging[i][1] == 'N' or  tagging[i][1] == 'P': #체언&용언
            word_list.append(tagging[i][0])

    return word_list
  except:
    return sent

#상담 데이터 품사 태깅
dsm['pos'] = dsm['word'].apply(extract_words)

dsm
```

우울장애 유사도

```python
# word2vec 모델 학습에 로그를 찍을 수 있도록 합니다.
import logging
logging.basicConfig(
    format='%(asctime)s : %(levelname)s : %(message)s', 
    level=logging.INFO)
```

```python
def to_list(sent):
  try:
    return eval(sent)
  except:
    return None

depressed['pos'] = depressed['pos'].apply(to_list)

depressed = depressed.dropna()

depressed['pos']
```

```python
from soynlp.tokenizer import RegexTokenizer

tokenizer = RegexTokenizer()

#토큰화
depressed['pos'] = depressed['pos'].apply(' '.join)

tokens = depressed['pos'].apply(tokenizer.tokenize)
depressed['tokenized'] = depressed['pos'].apply(tokenizer.tokenize)

depressed['tokenized']
```

```python
# 초기화 및 모델 학습
from gensim.models import word2vec

# 모델 학습
model = word2vec.Word2Vec(tokens, min_count=1)

# 모델 이름을 지정하고 저장한다.
model_name = 'depressed'
model.save(model_name)
```

```python
def similar(word):
  try:
    for w in word:
      words = []
      for i,j in model.wv.most_similar(w):
        if j > 0.9 and len(i)>1: #0.9을 threshold로 잡음
          words.append(i)
    return words
  except:
    return None

dsm['similar'] = dsm['pos'].apply(similar)
dsm
```

```python
def to_list(words):
  try:
    temp.extend(words)
  except: 
    pass
  return temp

df = {}

for i in range(1,10):
  temp = []
  temp2 = pd.DataFrame()
  temp2["temp"] = dsm[dsm["label"]==i]["similar"].apply(to_list)
  df[i] = set(temp2["temp"].iloc[0])

df
```

```python

lexicon = pd.DataFrame.from_dict(df, orient='index')
lexicon = lexicon.transpose()
lexicon
```
