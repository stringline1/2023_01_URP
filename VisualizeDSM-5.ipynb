#웰니스 데이터셋

```python
#웰니스 상담 데이터 합치기

df1 = pd.read_csv('/content/drive/MyDrive/학연생/웰니스-우울장애 표현/웰니스/웰니스 DSM-5 분류.csv', encoding='cp949', header=0, usecols = [1,2])
df2 = pd.read_csv('/content/drive/MyDrive/학연생/웰니스-우울장애 표현/웰니스/웰니스 DSM-5 분류2.csv', encoding='cp949', header=0, usecols = [1,2])
df3 = pd.read_csv('/content/drive/MyDrive/학연생/웰니스-우울장애 표현/웰니스/웰니스 DSM-5 분류3.csv', encoding='cp949', header=0, usecols = [1,2])
df4 = pd.read_csv('/content/drive/MyDrive/학연생/웰니스-우울장애 표현/웰니스/웰니스 DSM-5 분류4.csv', encoding='cp949', header=0, usecols = [1,2])
df5 = pd.read_csv('/content/drive/MyDrive/학연생/웰니스-우울장애 표현/웰니스/웰니스 DSM-5 분류5.csv', encoding='cp949', header=0, usecols = [1,2])
df6 = pd.read_csv('/content/drive/MyDrive/학연생/웰니스-우울장애 표현/웰니스/웰니스 DSM-5 분류6.csv', encoding='cp949', header=0, usecols = [1,2])
df7 = pd.read_csv('/content/drive/MyDrive/학연생/웰니스-우울장애 표현/웰니스/웰니스 DSM-5 분류7.csv', encoding='cp949', header=0, usecols = [1,2])
df8 = pd.read_csv('/content/drive/MyDrive/학연생/웰니스-우울장애 표현/웰니스/웰니스 DSM-5 분류8.csv', encoding='cp949', header=0, usecols = [1,2])
df9 = pd.read_csv('/content/drive/MyDrive/학연생/웰니스-우울장애 표현/웰니스/웰니스 DSM-5 분류9.csv', encoding='cp949', header=0, usecols = [1,2])

data = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9])
print(data)
```

```python
#데이터 중복 제거 (상담 데이터 column 기준), 인덱스 초기화
data = data.drop_duplicates(['상담 데이터'])
data = data.reset_index(drop=True)

data = data.drop(columns=['포함된 DSM-5 형태소'])
print(data)
```

```python
#문장, DSM-5 포함 단어, 레이블 포함한 well 데이터프레임 생성

well = pd.DataFrame(columns=['sentence', 'DSM-5','label'])

for sent in data['상담 데이터']:
  for i,j in zip(dsm[0], dsm['label']):
    if i in sent:
      well.loc[len(well.index)] = [sent, i, j]

print(well)
```

```python
label_freq = well.groupby(['label']).count() 
label_freq = label_freq.sort_values(by=['sentence'], ascending=False)
print(label_freq)

word_freq = well.groupby(['DSM-5']).count() 
word_freq = word_freq.sort_values(by=['sentence'], ascending=False)
print(word_freq)
```

```python
import matplotlib.pyplot as plt
import matplotlib

plt.rc('font', family='NanumBarunGothic') 

label_freq.plot.bar(y='DSM-5', rot=0)

plt.title("웰니스 DSM-5 빈도수")
plt.xlabel("카테고리")
plt.ylabel("빈도수")

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib

plt.rc('font', family='NanumBarunGothic') 

word_freq[:10].plot.bar(y='sentence', rot=0)

plt.title("웰니스 단어 빈도수")
plt.xlabel("단어")
plt.ylabel("빈도수")

plt.show()
```

#에브리타임 데이터셋

```python
#데이터 불러온 후 합치기

import pandas as pd

edata1 = pd.read_csv('/content/drive/MyDrive/학연생/에브리타임/everytime_data1.csv', encoding='UTF8', usecols = [1])
edata2 = pd.read_csv('/content/drive/MyDrive/학연생/에브리타임/everytime_data2.csv', encoding='UTF8', usecols = [1])
edata3 = pd.read_csv('/content/drive/MyDrive/학연생/에브리타임/everytime_data3.csv', encoding='UTF8', usecols = [1])

edata = pd.concat([edata1, edata2, edata3])
edata = edata.reset_index(drop=True)

print(edata)
```

```python
#문장, DSM-5 포함 단어, 레이블 포함한 well 데이터프레임 생성

every = pd.DataFrame(columns=['sentence', 'DSM-5','label'])

for sent in edata['article']:
  for i,j in zip(dsm[0], dsm['label']):
    if i in sent:
      every.loc[len(every.index)] = [sent, i, j]

print(every)
```

```python
label_freq = every.groupby(['label']).count() 
label_freq = label_freq.sort_values(by=['sentence'], ascending=False)
print(label_freq)

word_freq = every.groupby(['DSM-5']).count() 
word_freq = word_freq.sort_values(by=['sentence'], ascending=False)
print(word_freq)
```

```python
import matplotlib.pyplot as plt
import matplotlib

plt.rc('font', family='NanumBarunGothic') 

label_freq = label_freq.sort_values(by=['sentence'], ascending=False)
label_freq.plot.bar(y='DSM-5', rot=0)

plt.title("에브리타임 DSM-5 빈도수")
plt.xlabel("카테고리")
plt.ylabel("빈도수")

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib

plt.rc('font', family='NanumBarunGothic') 

word_freq[:10].plot.bar(y='sentence', rot=0)

plt.title("에브리타임 단어 빈도수")
plt.xlabel("단어")
plt.ylabel("빈도수")

plt.show()
```
