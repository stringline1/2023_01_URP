```python
import nltk
from konlpy.tag import Hannanum

han = Hannanum()

def extract_words(sent):
  tagging = han.pos(sent)
  word_list = []
  for i in range(0, len(tagging)):
      if tagging[i][1] == 'N': #명사
          word_list.append(tagging[i][0])
  return word_list
  
data['pos'] = data['상담 데이터'].apply(extract_words)
data
```

```python
import nltk
from konlpy.tag import Hannanum

han = Hannanum()

#품사 태깅
def extract_words(sent):
  tagging = han.pos(sent)
  word_list = []
  for i in range(0, len(tagging)):
      if tagging[i][1] == 'N' or  tagging[i][1] == 'P': #체언&용언
          word_list.append(tagging[i][0])
  return word_list

#상담 데이터 품사 태깅
data['pos'] = data['상담 데이터'].apply(extract_words)
data
```

```python
#인덱스 초기화
words = words.reset_index(drop=True)
words
```

```python
def labeling(datas):
  labels = []
  for i in range(len(words)):
    if set(words.iloc[i, 1])&set(datas) == set(words.iloc[i, 1]): #data안에 words가 포함되어 있는 경우
      labels.append(words.iloc[i,2])
  return labels

#문장에 존재하는 레이블들 추가
data['label'] = data['pos'].apply(labeling)
```

#웰니스 Sentence Transformer 분석

```python
!pip install sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('paraphrase-distilroberta-base-v1')

word_list = data["상담 데이터"].to_list()

embeddings = model.encode(word_list, convert_to_tensor=True)
cosine_scores = util.pytorch_cos_sim(embeddings, embeddings)
```

```python
org = 500
temp = cosine_scores[org]
temp.argsort(descending=False)[0:10]

for i in temp.argsort(descending=False)[0:10]:
    print(f"{i}. {word_list[org]} <> {word_list[i]} \nScore: {cosine_scores[org][i]:.4f}")
```

```python
#데이터 불러온 후 합치기

import pandas as pd

every1 = pd.read_csv('/content/drive/MyDrive/학연생/에브리타임/everytime_data1.csv', encoding='UTF8', usecols = [1])
every2 = pd.read_csv('/content/drive/MyDrive/학연생/에브리타임/everytime_data2.csv', encoding='UTF8', usecols = [1])
every3 = pd.read_csv('/content/drive/MyDrive/학연생/에브리타임/everytime_data3.csv', encoding='UTF8', usecols = [1])

every = pd.concat([every1, every2, every3])
every = every.reset_index(drop=True)

print(every)
```

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('paraphrase-distilroberta-base-v1')

word_list = every["article"].to_list()

embeddings = model.encode(word_list, convert_to_tensor=True)
cosine_scores = util.pytorch_cos_sim(embeddings, embeddings)
```

```python
org = 10000
temp = cosine_scores[org]
temp.argsort(descending=False)[0:10]

for i in temp.argsort(descending=False)[0:10]:
    print(f"{i}. {word_list[org]} <> {word_list[i]} \nScore: {cosine_scores[org][i]:.4f}")
```
